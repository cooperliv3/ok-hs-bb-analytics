```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/PAN/2025 Maxpreps Basketball Leaders - OK - Sheet1.csv')
df.head()
```

```python
#Count Number of Each Player Class in the Top 100 Scorers (PPG)
# Get total count of each class
counts = df['Class'].value_counts().reset_index()

# Rename columns to "Class" and "Count"
counts.columns = ['Class', 'Count']

# Print the DataFrame
counts.head()
```

```python
#Average Height of Each Class
# Function to convert "6'0" to inches
def height_to_inches(h):
    try:
        if isinstance(h, str):
            h = h.strip()
            feet, inches = h.split("'")
            return int(feet) * 12 + int(inches)
        else:
            return None
    except Exception:
        return None

# Define the order of the 'Class' column manually
class_order = ['Senior', 'Junior', 'Sophomore', 'Freshman']
df['Class'] = pd.Categorical(df['Class'], categories=class_order, ordered=True)

# Convert height and clean data
df['Height_in_inches'] = df['Height'].apply(height_to_inches)
df = df.dropna(subset=['Height_in_inches'])

# Group and calculate average height (no warning)
average_heights = df.groupby('Class', observed=True)['Height_in_inches'].mean().reset_index()

# Convert average back to feet and inches
average_heights['Feet'] = (average_heights['Height_in_inches'] // 12).astype(int)
average_heights['Inches'] = (average_heights['Height_in_inches'] % 12).round().astype(int)
average_heights['AVG Height'] = average_heights['Feet'].astype(str) + "'" + average_heights['Inches'].astype(str) + '"'

# Final output
average_heights = average_heights[['Class', 'AVG Height']]
average_heights = average_heights.sort_values('Class', ascending=True)

# Print
average_heights.head()
```
